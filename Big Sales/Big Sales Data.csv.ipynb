```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Big Sales Data.csv")
```

```python
df.head()
```

```python
df['Item_Fat_Content'] = df['Item_Fat_Content'].str.replace("LF", "Low Fat")
```

```python
df['Item_Fat_Content'] = df['Item_Fat_Content'].str.replace("low fat", "Low Fat")
```

```python
df['Item_Fat_Content'] = df['Item_Fat_Content'].str.replace("reg", "Regular")
```

```python
df.head()
```

```python
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)
```

```python
sns.countplot(data= df, x = "Item_Fat_Content", palette = 'pastel')
plt.title('Distribution of Item Fat Content')
plt.show()
```

```python
df.head(2)
```

```python
sns.scatterplot(data = df, x = "Item_Visibility", y = "Item_Outlet_Sales", hue = "Item_Fat_Content")
```

```python
sns.scatterplot(data=df, x='Item_MRP', y='Item_Outlet_Sales', hue='Outlet_Type', alpha=0.6)
plt.title('Item MRP vs Sales')
plt.show()
```

```python
sales_by_outlet = df.groupby('Outlet_Type')['Item_Outlet_Sales'].sum().reset_index()
sns.barplot(data=sales_by_outlet, x='Outlet_Type', y='Item_Outlet_Sales', palette='Set2')
plt.title('Total Sales by Outlet Type')
plt.xticks(rotation=45)
plt.show()
```

```python
df.head(2)
```

```python
sns.countplot(data=df, x='Outlet_Size', palette='Set1')
plt.title('Outlet Size Distribution')
plt.show()
```

```python
sns.boxplot(data=df, x='Outlet_Location_Type', y='Item_Outlet_Sales', palette='Accent')
plt.title('Sales by Outlet Location')
plt.show()
```

```python
df.head(3)
```

```python
top_item = df.groupby("Item_Type")["Item_Outlet_Sales"].sum().sort_values(ascending = False).reset_index()
sns.barplot(data = top_item, x = 'Item_Outlet_Sales', y = "Item_Type", palette = "cool" )
plt.title('Top 10 Item Types by Sales')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
corr = df.select_dtypes(include='number').corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
yearly_sales = df.groupby('Outlet_Establishment_Year')['Item_Outlet_Sales'].mean().reset_index()
sns.lineplot(data=yearly_sales, x='Outlet_Establishment_Year', y='Item_Outlet_Sales', marker='o')
plt.title('Avg Sales by Year of Establishment')
plt.show()
```

```python
sns.boxplot(data=df, x='Outlet_Type', y='Item_Outlet_Sales', palette='Spectral')
plt.title('Sales Distribution by Outlet Type')
plt.xticks(rotation=45)
plt.show()
```
