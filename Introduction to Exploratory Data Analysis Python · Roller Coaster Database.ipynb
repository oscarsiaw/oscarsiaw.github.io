```python
df  = pd.read_csv('coaster_db.csv')
```

```python
df[[
     #'coaster_name', 'Length', 'Speed', 
    'Location', 'Status', 
    #'Opening date',
       #'Type', 
    'Manufacturer', 
    #'Height restriction', 'Model', 'Height',
       #'Inversions', 'Lift/launch system', 'Cost', 'Trains', 'Park section',
      # 'Duration', 'Capacity', 'G-force', 'Designer', 'Max vertical angle',
     #  'Drop', 'Soft opening date', 'Fast Lane available', 'Replaced',
     #  'Closing date', 'Opened', 'Replaced by', 'Website',
     #  'Single rider line available', 'Restraint Style',
     #  'Flash Pass available', 'Acceleration', 'Restraints', 'Name',
     #  
    'year_introduced', 'latitude', 'longitude', 'Type_Main',
       'opening_date_clean', 
    #'speed1', 'speed2', 'speed1_value', 
    'speed1_unit',
       'speed_mph', 
    #'height_value', 'height_unit', 
    'height_ft',
       'Inversions_clean', 'Gforce_clean']],
```

```python

df = df[[
    
    'Location', 'Status', 
    'Manufacturer', 
    'year_introduced', 'latitude', 'longitude', 'Type_Main',
       'opening_date_clean', 
       'speed_mph', 
    'height_ft',
       'Inversions_clean', 'Gforce_clean']]
```

```python
df
```

```python
df.dtypes
```

```python
df['opening_date_clean'] = pd.to_datetime(df['opening_date_clean'])
```

```python
df.columns
```

```python
df = df.rename(columns = {'year_introduced': 'Year_Introduced',
                    'latitude': 'Latitude',
                     'opening_date_clean': 'Opening_Date_Clean',
                     'longitude': 'Longitude',
                     'speed_mph': 'Speed_Mph',
                     'height_ft': 'Height_ft',
                     'Inversions_clean': 'Inversions',
                    'Gforce_clean': 'Gforce'
                    })
```

```python
df.isna().sum()
```

```python
df = df.drop_duplicates()
```

```python
df = df.loc[~df.duplicated(subset = ['Location','Status', 'Manufacturer'  ])] \
.reset_index(drop = True).copy()
```

```python
sns.pairplot(df, vars = ["Year_Introduced", "Speed_Mph",
                         "Height_ft", "Inversions", "Gforce"],
                            hue = "Type_Main")
plt.show()
```

```python
df_corr = df[["Year_Introduced", "Speed_Mph",
    "Height_ft", "Inversions", "Gforce"]].dropna().corr()
df_corr
```

```python
ax = df.query('Location != "Other"') \
.groupby('Location')['Speed_Mph']\
.agg(['mean', 'count'])\
.query('count >= 10')\
.sort_values('mean')['mean']\
.plot(kind = 'barh', figsize=(12, 5), title = 'Average Coast Speed by Location')
ax.set_xlabel("Average Coaster Speed")
plt.show()
```
