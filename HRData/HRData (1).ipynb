```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
df = pd.read_csv('HRDataset.csv')
```

```python
df["DOB"] = df["DOB"].str.replace("-", "/")
```

```python
df.head()
```

```python
df['DOB'] =  pd.to_datetime(df['DOB'], errors = 'coerce')
```

```python
df["LastPerformanceReview_Date"] = pd.to_datetime(df["LastPerformanceReview_Date"], errors = 'coerce')
```

```python
df["LastPerformanceReview_Date"] = df["LastPerformanceReview_Date"].replace("-", "/")
```

```python
df["Gender"].value_counts()
```

```python
sns.countplot(data = df, x = 'Gender')
plt.title("Dender Distributions")
plt.tight_layout
plt.show()
```

```python
sns.countplot(data = df, x = 'MaritalDesc')
plt.title('Marital Status Distribution')
plt.xlabel("Marital Status")
plt.ylabel("Number of Employees")
plt.show()
```

```python
status_count = df["EmploymentStatus"].value_counts()
plt.pie(status_count, labels = status_count.index)
plt.title('Employment Status Distribution')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(7, 5))
sns.countplot(data=df, x='PerformanceScore', order=df['PerformanceScore'].value_counts().index, palette='Blues')
plt.title('Performance Score Distribution')
plt.xlabel('Performance Score')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['Absences'], bins=15, kde=True, color='salmon')
plt.title('Distribution of Absences')
plt.xlabel('Number of Absences')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
df.groupby('Gender')['Absences'].mean().plot(kind='bar', color='lightcoral', edgecolor='black')
plt.title('Average Absences by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Absences')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
sns.scatterplot(data=df, x='EngagementSurvey', y='EmpSatisfaction', hue='PerformanceScore')
plt.title('Engagement vs Satisfaction by Performance Score')
plt.xlabel('Engagement Survey')
plt.ylabel('Employee Satisfaction')
plt.legend(title='Performance Score', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
df['ManagerName'].value_counts().head(10).plot(kind='barh', color='mediumseagreen', edgecolor='black')
plt.title('Top 10 Managers by Team Size')
plt.xlabel('Number of Employees')
plt.ylabel('Manager')
plt.tight_layout()
plt.show()
```

```python
sns.boxplot(data=df, x='MaritalDesc', y='Absences', palette='pastel')
plt.title('Absences Distribution by Marital Status')
plt.xlabel('Marital Status')
plt.ylabel('Absences')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
