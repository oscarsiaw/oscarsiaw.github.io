```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Bike Prices.csv")
```

```python
df.head()
```

```python
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)
```

```python
print(df['Brand'])
```

```python
sns.histplot(df['Selling_Price'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Selling Price')
plt.xlabel('Selling Price')
plt.show()
```

```python
sns.scatterplot(data=df, x='Ex_Showroom_Price', y='Selling_Price', hue='Brand')
plt.title('Selling Price vs Ex-Showroom Price')
plt.show()
```

```python
sns.histplot(df['KM_Driven'], bins=30, kde=True, color='orange')
plt.title('Distribution of KM Driven')
plt.xlabel('KM Driven')
plt.show()
```

```python
df.head(1)
```

```python
brand_count = df['Brand'].value_counts().reset_index()
brand_count.columns = ['Brand', 'Count'] 

sns.barplot(data=brand_count, x='Brand', y='Count', palette='Set2')
plt.title('Number of Bikes by Brand')
plt.xlabel('Brand')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
sns.boxplot(data=df, x='Seller_Type', y='Selling_Price', palette='Accent')
plt.title('Selling Price by Seller Type')
plt.show()
```

```python
sns.countplot(data=df, x='Owner', palette='pastel')
plt.title('Distribution of Ownership Types')
plt.xticks(rotation=30)
plt.show()
```

```python
df_sorted = df.sort_values(by='Year', ascending=True)
sns.lineplot(data=df_sorted, x='Year', y='Selling_Price', ci=None, marker='o')
plt.title('Year vs Selling Price Trend')
plt.show()
```

```python
df_corr = df[['Selling_Price', 'KM_Driven', 'Ex_Showroom_Price', 'Year']].corr()
sns.heatmap(df_corr, annot=True, cmap='Blues')
plt.title('Correlation Heatmap')
plt.show()
```
