```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("MobilePriceRange.csv")
```

```python
df.head()
```

```python
sns.countplot(data=df, x ='PriceRange')
plt.title("Distribution of Mobile phone by price")
plt.show()
```

```python
sns.boxplot(data = df, x = 'BatteryPower', y = 'PriceRange')
plt.show()
```

```python
sns.scatterplot(data = df, y = 'InternalMemory', x = "PriceRange")
plt.title("Internal Memory vs PriceRange")
plt.show()
```

```python
sns.countplot(data = df, x = "4G", hue = 'PriceRange')
plt.title('4G Support by Price Range')
plt.show()
```

```python
sns.boxplot(data = df, x = 'PriceRange', y = "RAM")
plt.title('Ram distribution by price')
plt.show()
```

```python
sns.countplot(data = df, x = 'NoOfCores', hue = 'PriceRange' )
plt.title("CPU cores by price range")
plt.show()
```

```python
sns.jointplot(data = df, x = 'FrontCameraPx', y = 'RearCameraPX', kind= 'scatter')
```

```python
df['ScreenSize'] = df['ScreenHeight'] * df['ScreenWidth']
sns.boxplot(data=df, x='PriceRange', y='ScreenSize')
plt.title('Screen Size by Price Range')
plt.show()
```
