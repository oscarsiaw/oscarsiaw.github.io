```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Admission Chance.csv")
```

```python
sns.set(style = 'whitegrid')
plt.rcParams["figure.figsize"] = (8, 5)
```

```python
df.head()
```

```python
sns.histplot(data = df, x = 'GRE Score', bins = 20, kde = True, color = 'Skyblue')
plt.title("GRE Score Distribution")
plt.xlabel("GRE Score")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.histplot(df["CGPA"], kde = True, color = "Green", bins = 20)
plt.title("Distribution of CGPA Score")
plt.xlabel("CGPA")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.histplot(data = df, x = "Chance of Admit", bins = 20, kde = True, color = "Orange")
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
df.head()
```

```python
sns.regplot(x = "GRE Score", y = "Chance of Admit", data = df, scatter_kws={"alpha" :0.5})
plt.title("GRE Score vs Change of Admit")
plt.show()
```

```python
sns.regplot(data = df, x = "TOEFL Score", y = "Chance of Admit", scatter_kws={"alpha": 0.5})
plt.title("TOEFL Score vs Chance of Admit")
plt.show()
```

```python
sns.regplot(data = df, x = "CGPA", y = "Chance of Admit", scatter_kws = {"alpha":0.5})
plt.title("CGPA vs Chance of Admit")
plt.show()
```

```python
df.head()
```

```python
sns.boxplot(data = df, x = "Research", y = "Chance of Admit", palette='coolwarm')
plt.title("Research vs Chance of Admit")
plt.show()
```

```python
sns.boxplot(x='University Rating', y='GRE Score', data=df, palette='coolwarm')
plt.title('University Rating vs GRE Score')
plt.show()
```

```python
sns.pairplot(df[['GRE Score', 'TOEFL Score','University Rating', 'CGPA', "Research", 'Chance of Admit']], hue = "Research")
```

```python
sns.lmplot(x='GRE Score', y='Chance of Admit', hue='University Rating', data=df, aspect=1.5, palette='muted')
plt.title('GRE Score vs Chance of Admit (by University Rating)')
plt.show()
```
